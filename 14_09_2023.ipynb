```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score, confusion_matrix, accuracy_score
from pylab import rcParams
rcParams['figure.figsize'] = 8, 8
```

```python
data = pd.read_csv('/content/creditcard.csv')
```

```python
data.head()
```

```python
data.Class.value_counts()
```

```python
X = data.iloc[:,:-1]
X.shape
```

```python

Y = data.iloc[:,-1]
```

```python
Y.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0, stratify = Y)
```

```python
from matplotlib.pylab import NaN
Y = Y.fillna(0)
```

```python
X = X.fillna(0)
```

```python
classifier1 = RandomForestClassifier() # For GBM, use GradientBoostingClassifier()
grid_values = {'n_estimators':[10, 20], 'max_depth':[3, 5]}
classifier = GridSearchCV(classifier1, param_grid = grid_values, scoring = 'roc_auc', cv=5)
```

```python
X = X.fillna(X_train.mean())
```

```python
classifier.fit(X_train, Y_train)
```

train_preds = classifier.predict(X_train)
